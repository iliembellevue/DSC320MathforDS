1.Calculating Limits Numerically - First we import our libraries and define our limits f(x) and g(x) according to our worksheet.

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

def f_of_x(x):
    return (2*(x**3)) - (4*x) + 1

def g_of_x(x):
    return ((np.exp(1)**x)-1)/x
```

```python
delta_x = .001
x = np.arange(2, 3, delta_x)
a = np.arange(3.001, 4, delta_x)
e = f_of_x(x)
d = f_of_x(a)
print(max(e),min(d))
```

```python
plt.plot(x,e)
plt.plot(a,d)
plt.show()
```

1(a).For f(x) as the move closer to the limit of 3, we approach 43


```python
delta_x = .01
x = np.arange(-1, 0, delta_x)
a = np.arange(0.01, 1, delta_x)
e = g_of_x(x)
d = g_of_x(a)
print(max(e),min(d))
```

```python
plt.plot(x,e)
plt.plot(a,d)
plt.show()
```

1(b).Using the function in numpy np.exp(1) in place of e, also known as Euler's number, we can calulate the limits as it moves closer to 0 which is 1. 

1(c).You can plug in the input into the first function f(x), but not into g(x), because g(x) is undefined at (0,1) coordinates. There is no break in the function f(x) so it is in fact a function of continuity. However, g(x) is not continuous because there is a break in the line at y=1 because of the undefined value of the function. 

2.Average Rate of Change

```python
def AROC(a, b):
    return ((3*(b**2)) - (3*(a**2)))/(b-a)

a = 5
b = 10
print("The Average Rate of Change with the given input is",AROC(a,b))
```

3.Average Rate of Change to Instantaneous Rate of Change

```python
def traveled_distance(t):
    return 4.9*(t**2)

def AROC2(a, b):
    return (traveled_distance(b)-traveled_distance(a))/(b-a)
```

```python
delta_t = 1
speed = np.arange(0, 7, delta_t)
distance = traveled_distance(speed)
print("3(a) The speed between 5 and 6 seconds is", round(max(distance) - distance[len(distance)-2],2), "meters per second.")
print("Using the Average Rate of Change formula, it is", round(AROC2(max(speed), speed[len(speed)-2]), 2), "meters per second.")
plt.scatter(max(speed), max(distance), color="red")
plt.scatter(speed[len(speed)-2], distance[len(distance)-2], color="red")
plt.plot(speed, distance)
plt.show()
```

```python
delta_t = .5
speed = np.arange(0, 6, delta_t)
distance = traveled_distance(speed)
print("3(b) Distance traveled between 5 and 5.5 seconds is", round((max(distance)-distance[len(distance)-2])/.5, 2), 
      "meters per second.")
print("Using the Average Rate of Change formula, it is", round(AROC2(max(speed), speed[len(speed)-2]), 2), "meters per second.")
plt.scatter(max(speed), max(distance), color="red")
plt.scatter(speed[len(speed)-2], distance[len(distance)-2], color="red")
plt.plot(speed, distance)
plt.show()
```

```python
delta_t = .1
speed = np.arange(0, 5.2, delta_t)
distance = traveled_distance(speed)
print("3(c) The speed between 5 and 5.1 seconds is", round((max(distance)-distance[len(distance)-2])/.1, 2),
      "meters per second.")
print("Using the Average Rate of Change formula, it is", round(AROC2(max(speed), speed[len(speed)-2]), 2), "meters per second.")
plt.scatter(max(speed), max(distance), color="red")
plt.scatter(speed[len(speed)-2], distance[len(distance)-2], color="red")
plt.plot(speed, distance)
plt.show()
```

4.Calculating and Interpreting Partial Derivatives

```python
def selling_price(c,y):
    return 16000 + (2400*c) - (1800*y)

p = selling_price(8,5)
print("4(a) The model's predicted selling price is",'${:.2f}'.format(p))
```

```python
print("4(b) When all other variables stay constant, but condition variable changes, the output of the partial derivative is",
      '${:.2f}'.format(p/2400))
```

```python
print("4(c) When all other variables stay constant, but year variable changes, the output of the partial derivative is",
      '${:.2f}'.format(p/1800))
```
