# Working with Matrices

```python
import numpy as np

# 1. Create A, B, and C matrices with numpy.
A = np.matrix([[1, -2, 3, 7],
              [2, 1, 1, 4],
              [-3, 2, -2, 10]])

B = np.matrix([[0, 1, -3, -2],
              [10, -1, 2, -3],
              [5, 1, -1, 4]])

C = np.matrix([[4, 0, -2, 3],
              [3, 6, 9, 7],
              [2, 2, 5, 1],
              [9, 4, 6, -2]])

# Output the dimensions of each matrix using .shape function.
print("2. The shape of A is:", A.shape, "\nB is:", B.shape, "\nC is:", C.shape)
```

```python
# Calculate the sum of A+B and display the matrix
print("3a) Is defined and can be calcuted by A + B:\n", A + B)
```

```python
# A * B produces an error so I displayed why.
print("3b) A * B cannot be defined since there is a different amount of columns to rows.")
```

```python
# Calculate the product of A*C and display the matrix
print("3c) Is defined and can be calcuted by A * C:\n", A*C)
```

```python
# Use the transposition function to pivot the tables to rows
print("3d) Is defined and can be calcuted by by using the transposition function:\n Original:\n", C, "\nTransposed:\n", C.T)
```

```python
# Display the properites of matrix tranpositions by calculating both ways.
print("4. Option A:\n", (A + B).T, "\n Option B:\n", A.T+B.T)
```

```python
# Display the properites of matrix tranpositions by calculating both ways.
print("5. Option A.\n", (A*C).T, "\n Option B:\n", (C.T * A.T))
```

```python
# Use the inverse function to find the inverse of the square matrix C.
print("6.", np.linalg.inv(C))
```
