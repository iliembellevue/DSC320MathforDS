1. Import the dataset and create a scatter plot with matplotlib

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

df = pd.read_csv("C:\\Users\ILiem\\OneDrive - The Scoular Company\\Documents\\Bellevue\\Math for DS\\week2data\\week2data\\car_data.csv")
df.plot.scatter(x='weight', y='hwy_mpg')
plt.show()
```

2.As weight increases, mpg decreases based upon the scatter plot above.

```python
from scipy import stats
slope, intercept, rvalue, p, se = stats.linregress(df['weight'], df['hwy_mpg'])
print('slope: {:.4f}'.format(slope))
print('intercept: {:.4f}'.format(intercept))
print('correlation coefficient: {:.4f}'.format(rvalue))
print('p-value: {:.4f}'.format(p))
print('std error: {:.4f}'.format(se))
```

3.The slope is negative based upon the linear regression model so it is indeed negative. This problem was easy to see with the eye, but it's always best to check.
4.If the slope of the line was -0.05 then there is a negative linear relationship. As one variable increases the other decreases.

```python
def f(x, slope, intercept):
    slope = slope
    intercept = intercept
    return slope * x + intercept

my_value = 3200
print('Predicted MPG: {:.2f}'.format(f(my_value, slope, intercept)))
```
