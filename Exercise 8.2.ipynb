```python
#Import the necessary libraries.
import numpy as np
import matplotlib.pyplot as plt

#Create an array matrix of the linear equation in the weekly lesson.
A = np.array([[27,-10,4,-29],
              [-16,5,-2,18],
              [-17,4,-2,20],
              [-7,2,1,8]])
print(A)
```

```python
#Create an array matrix of the solution side.
B = np.array([1,-1,0,1])

#Using numpy's solve function to find x in the lesson assignment.
x = np.linalg.solve(A,B)
print(x)
```

```python
#Check that the solution is correct by transforming back into B.
X = A @ x
print(X)
```

```python
#Create the 2x2 array matrix in the lesson assignment.
A = np.array([[np.sqrt(2)/2, -np.sqrt(2)/2],
               [np.sqrt(2)/2, np.sqrt(2)/2]])

#2(a) Create x according to the lesson assignment.
v = np.array([2,-2])

#Calculate the dot product of Av. 
Av = A @ v
print(Av)
```

```python
# Mimic the code in section 7.1.3 to create a scatter plot with color coordination.
x = np.arange(-10, 10, 1)
y = np.arange(-10, 10, 1)
xx, yy = np.meshgrid(x, y)
plt.scatter(xx, yy, s=20, c=xx+yy)
plt.show()
```

```python
#Structure the points of the grid to apply the matrix.
xy = np.vstack([xx.flatten(), yy.flatten()])
xy.shape
```

```python
#Apply A to each point to transform each of the cordinates of the vector.
trans = A @ xy
trans.shape
```

```python
#Reshape the arrays to have a similliar shape to xx and yy.
xx_transformed = trans[0].reshape(xx.shape)
yy_transformed = trans[1].reshape(yy.shape)
```

```python
#Show the new transformed vectors.
plt.scatter(xx_transformed,yy_transformed,s=10,c=xx+yy)
plt.show()
```
